```julia
using CairoMakie
```

```julia
x = range(0, 10, 100);
y_sin = sin.(x);
y_cos = cos.(x);
```

```julia
fig, ax = scatter(x, y_sin, color = :red, label = "sin",
                  figure = (; resolution = (800, 500)),
                  axis = (title = "ScatterPlots",
                          xlabel = "X Axis",
                          ylabel = "Sine/Cosine of X"))
scatter!(ax, x, y_cos, color = :blue, label = "cos")
axislegend()
```

```julia
current_figure()
```

```julia
fig = Figure(resolution = (1000, 500))
lines(fig[1, 1], x, y_sin, color = :red,
      axis = (; title = "sin(x)"))
lines(fig[1, 2], x, y_cos, color = :blue,
      axis = (; title = "cos(x)"))
lines(fig[2, :], x, tan.(x), color = :green, 
      axis = (; title = "tan(x)"))
current_figure()
```
