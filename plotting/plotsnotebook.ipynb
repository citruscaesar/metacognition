```julia
using CairoMakie
```

```julia
x = range(0, 10, 100);
y_sin = sin.(x);
y_cos = cos.(x);
```

```julia
fig, ax = scatter(x, y_sin,
                  figure = (; resolution = (800, 500)),
                  axis = (title = "ScatterPlots",
                          xlabel = "X Axis",
                          ylabel = "Sine/Cosine of X"))
scatter!(ax, x, y_cos)
```
