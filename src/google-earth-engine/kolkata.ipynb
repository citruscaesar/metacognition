```python
BOUNDS = {
    "top-left": [22.67141043215552, 88.21453658006386],
    "bottom-right": [22.442886523274655, 88.49095586110563],
    "rect": [88, 22.3, 88.7, 23]
}
```

```python
DATA_SOURCES = {
    "population": "JRC/GHSL/P2023A/GHS_POP",
    "yearly_water": "JRC/GSW1_4/GlobalSurfaceWater"
}
```

```python
pop_1975 = ee.Image(DATA_SOURCES["population"] + "/1975")
pop_1980 = ee.Image(DATA_SOURCES["population"] + "/1980")
pop_1985 = ee.Image(DATA_SOURCES["population"] + "/1985")
pop_1990 = ee.Image(DATA_SOURCES["population"] + "/1990")
pop_1995 = ee.Image(DATA_SOURCES["population"] + "/1995")
pop_2000 = ee.Image(DATA_SOURCES["population"] + "/2000")
pop_2005 = ee.Image(DATA_SOURCES["population"] + "/2005")
pop_2010 = ee.Image(DATA_SOURCES["population"] + "/2010")
pop_2015 = ee.Image(DATA_SOURCES["population"] + "/2015")
pop_2020 = ee.Image(DATA_SOURCES["population"] + "/2020")
pop_2025 = ee.Image(DATA_SOURCES["population"] + "/2025")
pop_2030 = ee.Image(DATA_SOURCES["population"] + "/2030")
pop_viz_params = {"min": 0.0, "max": 1000.0, "palette": ['060606', '337663', '337663', 'ffffff']}
```

```python
m = geemap.Map(center = [22.57, 88.36], zoom=8, height=600)
water = ee.Image(DATA_SOURCES["yearly_water"])
water_viz_params = {
  "bands": ['recurrence'],
  "min": 0.0,
  "max": 100.0,
  "palette": ['ffffff', 'ffbbbb', '0000ff']
};
m.add_layer(pop_2005, pop_viz_params, "GHS")
m.add_layer(water, water_viz_params, "Water")
#aoi = ee.Geometry.Rectangle(BOUNDS["rect"])
m
```
