```python
import ee
import geemap.foliumap as geemap
import matplotlib.pyplot as plt
%matplotlib widget

geemap.ee_initialize()
```

```python
BOUNDS = {
    "top-left": [22.67141043215552, 88.21453658006386],
    "bottom-right": [22.442886523274655, 88.49095586110563],
    "rect": [88, 22.3, 88.7, 23]
}
```

```python
DATA_SOURCES = {
    "population": "JRC/GHSL/P2023A/GHS_POP",
    "water": "JRC/GSW1_4/GlobalSurfaceWater",
    "flood_events": "GLOBAL_FLOOD_DB/MODIS_EVENTS/V1"
}
```

```python
m = geemap.Map(center = [22.57, 88.36], zoom=8, height=600)
flood = ee.ImageCollection(DATA_SOURCES["flood_events"]).filterMetadata("dfo_country", "equals", "India").select("flooded").sum()
flood_viz_params = {"min": 0, "max": 30, "palette": ['c3effe', '1341e8', '051cb0', '001133']}
m.add_layer(flood, flood_viz_params, "Flood")
m
```

```python
transition_palette = ["#ffffff",# 	No change (Transparent / White)
           "#0000ff",# 	Permanent (Blue)
           "#22b14c",# 	New permanent (Green)
           "#d1102d",# 	Lost permanent (Red)
           "#99d9ea",# 	Seasonal (Teal)
           "#b5e61d",# 	New seasonal (Light Green)
           "#e6a1aa",# 	Lost seasonal (Brown)
           "#ff7f27",# 	Seasonal to permanent (Orange)
           "#ffc90e",# 	Permanent to seasonal (Mustard)
           "#7f7f7f",# 	Ephemeral permanent (Light Gray)
           "#c3c3c3" #  Ephemeral seasonal (Dark Gray)
]

water = ee.Image(DATA_SOURCES["water"])
water_viz_params = {
  "bands": ['transition'],
  "min": 0,
  "max": 10,
  "palette": transition_palette
};

m = geemap.Map(center = [22.57, 88.36], zoom=8, height=600)
m.add_layer(water, water_viz_params, "Water Transition")
m
```
