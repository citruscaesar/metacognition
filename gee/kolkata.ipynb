```python
import ee
import geemap.foliumap as geemap
import matplotlib.pyplot as plt
#%matplotlib widget

geemap.ee_initialize()
```

```python
region = ee.Geometry.BBox(88.014709, 21.5394460000001, 89.006367, 23.2563400280001)
roi = ee.FeatureCollection(region)
roi_viz_params = {'color': 'ffff00ff', 'fillColor': '00000000'}

water_history = ee.Image("JRC/GSW1_4/MonthlyHistory/2016_01").select("water")
water_viz_params = {"min": 0.0, "max": 2.0, "palette": ['ffffff', 'fffcb8', '0905ff']}

m = geemap.Map()#center = [22.57, 88.36], zoom=8, height=600)
m.center_object(roi, 8)
m.add_layer(water_history, water_viz_params)
m.add_layer(roi, roi_viz_params)
m
```

```python
for year in range(2001, 2015):
    for month in range(1, 12):
        date_str = f"{year}_{month:02d}"
        water_history = ee.Image(f"JRC/GSW1_4/MonthlyHistory/{date_str}").select("water")
        geemap.ee_export_image(water_history, filename=f"sundarbans_flood_inundation_{date_str}.tif", scale=30, region=region)
```

```python
transition_palette = ["#ffffff",# 	No change (Transparent / White)
           "#0000ff",# 	Permanent (Blue)
           "#22b14c",# 	New permanent (Green)
           "#d1102d",# 	Lost permanent (Red)
           "#99d9ea",# 	Seasonal (Teal)
           "#b5e61d",# 	New seasonal (Light Green)
           "#e6a1aa",# 	Lost seasonal (Brown)
           "#ff7f27",# 	Seasonal to permanent (Orange)
           "#ffc90e",# 	Permanent to seasonal (Mustard)
           "#7f7f7f",# 	Ephemeral permanent (Light Gray)
           "#c3c3c3" #  Ephemeral seasonal (Dark Gray)
]

water = ee.Image(DATA_SOURCES["water"])
water_viz_params = {
  "bands": ['transition'],
  "min": 0,
  "max": 10,
  "palette": transition_palette
};

m = geemap.Map(center = [22.57, 88.36], zoom=8, height=600)
m.add_layer(water, water_viz_params, "Water Transition")
m
```
